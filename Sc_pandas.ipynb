```python
import pandas as pd
import numpy as np
```

```python
df_sc_pandas = pd.read_csv("https://raw.githubusercontent.com/LuisaPolicarpo/Project_6/main/Datasets/socio_context.csv")
```

```python
df_sc_pandas_du = pd.get_dummies(df_sc_pandas, columns=['Variable'])
```

```python
df_sc_pandas_du.columns
```

```python
df_sc_pandas_final= df_sc_pandas_du.drop(columns=["COU", 'YEA','Unit Code', 'PowerCode Code', 'PowerCode', 'Reference Period Code', 'Flag Codes', 'Flags',
       'Variable_GDP deflator', 'Variable_Labour tax revenue, % GDP',
       'Variable_Labour tax revenue, % total tax revenue', 'Variable_Nominal exchange rate', 'Variable_Purchasing power parity','Variable_Real GDP, Index 2000=100'])
```

```python
df_sc_pandas_final.columns
```

```python
df_sc_pandas_final['Variable_Population, ages 65 and above, % total'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Population, ages 65 and above, % total'])
```

```python
df_sc_pandas_final['Variable_Life expectancy at birth'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Life expectancy at birth'])
```

```python
df_sc_pandas_final['Variable_Net migration'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Net migration'])
```

```python
df_sc_pandas_final['Variable_Population'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Population'])
```

```python
df_sc_pandas_final['Variable_Population density, inhabitants per km2'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Population density, inhabitants per km2'])
```

```python
df_sc_pandas_final['Variable_Population, ages 0-14, % total'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Population, ages 0-14, % total'])
```

```python
df_sc_pandas_final['Variable_Population, ages 15-64, % total'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Population, ages 15-64, % total'])
```

```python
df_sc_pandas_final['Variable_Real GDP per capita'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Real GDP per capita'])
```

```python
df_sc_pandas_final['Variable_Total fertility rate, children per woman'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Total fertility rate, children per woman'])
```

```python
df_sc_pandas_final['Variable_Value added in agriculture, % of total value added'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Value added in agriculture, % of total value added'])
```

```python
df_sc_pandas_final['Variable_Value added in industry, % of total value added'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Value added in industry, % of total value added'])
```

```python
df_sc_pandas_final['Variable_Value added in services, % of total value added'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Value added in services, % of total value added'])
```

```python
df_sc_pandas_final['Variable_Women, % total population'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Women, % total population'])
```

```python
df_sc_pandas_final['Variable_Population, ages 65 and above, % total_2'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['Variable_Population, ages 65 and above, % total_2'])
```

```python
df_sc_pandas_final['\tVariable_Life expectancy at birth'] = (df_sc_pandas_final['Value']*df_sc_pandas_final['\tVariable_Life expectancy at birth'])
```
